```python
def accuracy(A,data,label):
    pred = np.dot(np.linalg.pinv(A),data)
    pred=np.dot(B,pred).T
    row_sums = pred.sum(axis=1)
    pred = pred / row_sums[:, np.newaxis]
    print(pred)
    pred=np.argmax(pred, axis=1)
    print(pred)
    accuracy=np.equal(pred,label).sum()/label.shape[1]
    return accuracy
```

```python
def ssnmf(X,y,lamda = 0.75, r = 35,n_itr = 1000, k=4):
    m, n = X.shape
    y = y.reshape(-1)
    print(m,n)
    S = np.random.rand(r, n)
    A = np.random.rand(m, r)
    B = np.random.rand(k, r)
    Y = np.eye(k)[y].T
    eps = np.spacing(1)
    try:
        with open('variable.pkl', 'rb') as f:
            A = pickle.load(f)
            B = pickle.load(f)
            S = pickle.load(f)
            f.close()
    except Exception:
        S = np.random.rand(r, n)
        A = np.random.rand(m, r)
        B = np.random.rand(k, r)
        
    for i in range(n_itr):
        A *= (np.dot( X, S.T)) / (np.dot(np.dot( A, S), S.T) + eps)
        B *= (np.dot( Y, S.T)) / (np.dot(np.dot( B, S), S.T) + eps)
        S *= (np.dot( A.T, X)+ lamda * np.dot( B.T, Y)) / (np.dot( A.T, np.dot( A, S)+ eps)+ lamda * np.dot( B.T, np.dot( B, S)+ eps))
        print("iteration", i, euclid_divergence(X, Y, A, B, S, lamda ))
    
    with open('variable.pkl', 'wb') as f:
        pickle.dump(A, f)
        pickle.dump(B, f)
        pickle.dump(S, f)
        f.close()
    return S, A, B    
        

def euclid_divergence( X, Y, A, B, S, lamda):
    d = ((X - np.dot(A, S))**2).sum() + lamda * ((Y - np.dot(B, S))**2).sum() 
    return d        
```

```python
def ssnmfnew(X,y,lamda = 0.75, r = 35,n_itr = 1000, k=4):
    m, n = X.shape
    y = y[:,0:300]
    y = y.reshape(-1)
    L = np.ones([k,300])
    L_test = np.zeros([k,80])
    L = np.concatenate((L,L_test),axis=1)
    y_test = np.zeros([k,80])
    print(m,n)
    S = np.random.rand(r, n)
    A = np.random.rand(m, r)
    B = np.random.rand(k, r)
    Y = np.eye(k)[y].T
    Y = np.concatenate((Y,y_test),axis=1)
    eps = np.spacing(1)
    try:
        with open('variable_methyl.pkl', 'rb') as f:
            A = pickle.load(f)
            B = pickle.load(f)
            S = pickle.load(f)
            f.close()
    except Exception:
        S = np.random.rand(r, n)
        A = np.random.rand(m, r)
        B = np.random.rand(k, r)
        
    for i in range(n_itr):
        A *= (np.dot( X, S.T)) / (np.dot(np.dot( A, S), S.T) + eps)
        B *= (np.dot(L * Y, S.T)) / (np.dot(L * np.dot( B, S), S.T) + eps)
        S *= (np.dot( A.T, X)+ lamda * np.dot( B.T,L * Y)) / (np.dot( A.T, np.dot( A, S)+ eps)+ lamda * np.dot( B.T,L * np.dot( B, S)+ eps))
        print("iteration", i, euclid_divergence(X, Y, A, B, S, lamda ))
    
    with open('variable_methyl.pkl', 'wb') as f:
        pickle.dump(A, f)
        pickle.dump(B, f)
        pickle.dump(S, f)
        f.close()
    return S, A, B    
```
